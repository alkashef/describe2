```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
```

```python
def unique(series):
    return len(series.unique())

def count0(series):
    frequencies = series.value_counts().to_dict()
    return frequencies.get(0, 0)

def count_positive(series):
    return sum(i > 0 for i in series.values)

def count_negative(series):
    return sum(i < 0 for i in series.values)

columns = ['Type',
               'Density'
               'No. of Values', 
               'No. of Unique Values', 
               'No. of NaN',
               'No. of Zeros',
               'No. of +ve Value',
               'No. of -ve Value',
               'Min',
               'Max',
               'Mean',
               'Std',
               'Median', 
               'Mode',
               'Skewness',
               'Kurtosis']
    
def describe2(df):
    df = df.select_dtypes(exclude=['object', 'bool'])
    
    description = pd.DataFrame()
    
    description['Type'], description['Density'] = df.ftypes.str.split(':').str
    description['No. of Values'] = df.apply(len, axis=0)
    description['No. of Unique Values'] = df.apply(unique, axis=0)
    description['No. of NaN'] = description['No. of Values'] - df.count(axis=0)
    description['No. of Zeros'] = df.apply(count0, axis=0)
    description['No. of +ve Value'] = df.apply(count_positive, axis=0)
    description['No. of -ve Value'] = df.apply(count_negative, axis=0)
    
    return description

describe2(df)
```
